```python
import os

import numpy as np
import pandas as pd

import cv2 
import PIL as pil
from PIL import Image

from IPython.display import display
import matplotlib.pyplot as plt
```

```python
def get_data_path(train=True):
    if train:
        return "resources/MSRA-TD500/train"
    return "resources/MSRA-TD500/test"
```

```python
def get_images_dataset():
    path = os.path.abspath(get_data_path())
    images_paths = [file for file in os.listdir(path) 
                   if "jpg" in file.lower()]
    images_abs_paths = [os.path.join(path, file) for file in images_paths]
    images_ids = [file[:-4] for file in images_paths]
    dataset = pd.DataFrame({"imagepath" : images_abs_paths,
                            "imageid": images_ids})
    return dataset
```

```python
def show_image(imagepath):
    im = Image.open(imagepath)
    plt.figure(figsize=(5, 5))
    plt.imshow(im)
    plt.show()
```

```python
def get_sample_image(dataset):
    _id = np.random.choice(dataset.shape[0])
    return dataset.loc[_id]["imagepath"]
```

```python
data = get_images_dataset()
```

```python
data.head()
```

```python
show_image(get_sample_image(data))
```

```python
def apply_binary_filter(imgpath):
    img = cv2.imread(imgpath)
    ret,thresh1 = cv2.threshold(img, 127, 255.0, cv2.THRESH_BINARY)
    return thresh1
```

```python
def apply_grayscale_filter(imgpath):
    img = cv2.imread(imgpath)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img_gray
```

```python
imgpath = get_sample_image(data)

show_image(imgpath)

filtered_img = apply_grayscale_filter(imgpath)
plt.imshow(filtered_img, cmap="gray")
plt.show()
```

```python
imgpath = get_sample_image(data)

show_image(imgpath)

filtered_img = apply_binary_filter(imgpath)
plt.imshow(filtered_img)
plt.show()
```
